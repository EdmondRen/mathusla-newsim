```python
from pathlib import Path
import os, sys
from importlib import reload
import pprint

sys.path.append("../../../python")

import matplotlib.pyplot as plt
import numpy as np

import simhelper.root as iroot
from simhelper import generator
reload(generator)
```

```python
# -------------------------------
# Read the filereader file
def load_old_filereader(filename):
    """
    Return:
    list of [[x, y, z, px, py, pz],[x, y, z, px, py, pz],...]
    """
    events = []
    current_event = None
    
    with open(filename, 'r') as file:
        for line in file:
            line = line.strip()
            
            # Ignore comments and blank lines
            if not line or line.startswith('#'):
                continue
            
            # New event
            if line.startswith('n'):
                if current_event is not None:
                    events.append(current_event)
                current_event = []
            else:
                # Parse and add data to the current event
                data = list(map(float, line.split()))
                current_event.append(data)
    
    # Append last event if it exists
    if current_event is not None:
        events.append(current_event)
    
    return events


def list_to_root(event_list, filename_output):
    #-----------------------------
    # setup output file
    if not os.path.exists(os.path.dirname(filename_output)):
        Path(os.path.dirname(filename_output)).mkdir( parents=True, exist_ok=True )
    tree_writer = iroot.tfile_writer("data", filename_output)
    tree_writer.define_branch("Gen_pdgID", 'vector<double>')    
    tree_writer.define_branch("Gen_x", 'vector<float>')    
    tree_writer.define_branch("Gen_y", 'vector<float>')    
    tree_writer.define_branch("Gen_z", 'vector<float>')    
    tree_writer.define_branch("Gen_t", 'vector<float>')    
    tree_writer.define_branch("Gen_px", 'vector<float>')    
    tree_writer.define_branch("Gen_py", 'vector<float>')    
    tree_writer.define_branch("Gen_pz", 'vector<float>')    

    keys = ["Gen_pdgID","Gen_x","Gen_y","Gen_z","Gen_t","Gen_px","Gen_py","Gen_pz"]

    for event in event_list:
        data = {key: [] for key in keys}
        for decay in event:
            data["Gen_pdgID"].append(decay[0])
            data["Gen_x"].append(decay[1]-90000)
            data["Gen_y"].append(decay[2])
            data["Gen_z"].append(-decay[3])
            data["Gen_t"].append(0)
            data["Gen_px"].append(decay[1+3])
            data["Gen_py"].append(decay[2+3])
            data["Gen_pz"].append(-decay[3+3])
        tree_writer.fill(data)


    tree_writer.write_and_close()    
```

```python
# Convert all files to new format
for f in llp_file_list:
    
    filename=f"{basedir}/{f}"
    events =  load_old_filereader(filename)
    
    filename_root = str(Path(filename).with_suffix(".root"))
    list_to_root(events, filename_root)
```
